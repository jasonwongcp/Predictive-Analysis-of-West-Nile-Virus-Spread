By uisng df.info(), I know the datatype of the columns.

```python
#Part 1-4
# Are there any columns that contain duplicate information? If so, remove the redundant columns.
df.drop_duplicates()
```

```python
# Part 1-5 Are there any null values in the dataframe? If so, deal with them appropriately.
for column in ["Year", "Week", "Address Block", "Block", "Trap", "Trap type","Date","Mosquito number", "Mosquito ID", "WNV Present","Species","Lat","Lon"]:
    df[column] = df[column].fillna("unknown")
```

We use the 'unknown` category to fill in the missing values.

```python
# Part 2 - Exploring the relationship between mosquito number and date
# Extract the 
df["Month"] = df["Date"].dt.month
df["Month"].info()
```

```python
plt.bar(df["Month"].value_counts().index, df["Month"].value_counts().values)
plt.xlabel('Date')
plt.ylabel('Number of Mosquito')
plt.title('Number of Mosquito over time')
plt.legend(loc='center right') # <- this surfaces the labels in a legend
```

When we show the number of mosquito by month, we can see that there are more mosquitoes during the summer time, i.e., July and August.

```python
# Part 3-1 Using an appropriate visual, explore the relationship between mosquito species and WNV prevalence.

```

```python
Prevalence = df[(df['WNV Present']=='positive')].reset_index(drop=True)

plt.bar(Prevalence['Species'].value_counts().index, Prevalence['Species'].value_counts().values)
plt.xlabel('Species')
plt.ylabel('Prevalence')
plt.title('Prevalence of WNV species')
plt.xticks(rotation=45)
plt.show()
```

When we categorize the 4 species of mosqutoes where WNV is positive, we know that CULEX REDTUANS is the most prevalent species while CULEX TERRITANS and CULEX SALINARIUS are the least prevalent.

```python
# Part 3-2 Using an appropriate visual, explore the relationship between mosquito species and WNV prevalence.

sns.boxplot(x="Trap type", y="Mosquito number", data=df, color="blue"
           )
plt.xlabel("Trap Type")
plt.ylabel("Number of Mosquito caught")
plt.title("Number of mosquito caught per Trap type")
plt.show()
```

If we use a boxplot to show the distribution of mosquito number cauhgt by different traps, we can learn that SENTINEL is able to catch the highest median of mosquito caught while OVI is not able to do do a good job in catching mosquitoes.

```python
# Part 3-3 Using an appropriate visual, come up with an additional insight of your choice.
plt.hist( df["Week"].values, bins=100)
plt.xlabel('Week')
plt.ylabel('Number of Mosquito')
plt.title('Number of Mosquito over time')
plt.axhline(y = 1000, color = 'r', linestyle = '-')
plt.legend(loc='center right') # <- this surfaces the labels in a legend
```
